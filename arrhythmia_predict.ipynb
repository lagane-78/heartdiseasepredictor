```python
import pandas as pd
import numpy as np
```

# 1. Full dataset with various classifiers.

```python
df = pd.read_csv('./kaggle-ArrythmiaData/arrhythmia.csv',encoding = "ISO-8859-1")
```

```python
# Fill Null values in some columns with mean value for the column
df['J'] = df['J'].fillna(df['J'].mean())
df['P'] = df['P'].fillna(df['P'].mean())
df['T'] = df['T'].fillna(df['T'].mean())
df['QRST'] = df['QRST'].fillna(df['QRST'].mean())
df['S1 Wave'] = df['S1 Wave'].fillna(df['S1 Wave'].mean())
```

```python
df
```

```python
# Split X and Y from CSV
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
np.where(np.isnan(X))

# Split 20% of the data for Testing Set randomly
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)


```

## 1.1 Logistic Regression Classifier

```python

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm_test = confusion_matrix(y_pred, y_test)

y_pred_train = classifier.predict(X_train)
cm_train = confusion_matrix(y_pred_train, y_train)

print()
print('Accuracy for training set for Logistic Regression = {}'.format((cm_train[0][0] + cm_train[1][1])/len(y_train)))
print('Accuracy for test set for Logistic Regression = {}'.format((cm_test[0][0] + cm_test[1][1])/len(y_test)))
```

## 1.2 KNN with 13 neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=7)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm_test = confusion_matrix(y_pred, y_test)

y_pred_train = classifier.predict(X_train)
cm_train = confusion_matrix(y_pred_train, y_train)

print()
print('Accuracy for training set for KNN = {}'.format((cm_train[0][0] + cm_train[1][1])/len(y_train)))
print('Accuracy for test test for KNN = {}'.format((cm_test[0][0] + cm_test[1][1])/len(y_test)))
```

## 1.3  SVM classifier

```python
from sklearn import svm
classifier = svm.SVC(C=1,kernel="linear")
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm_test = confusion_matrix(y_pred, y_test)

y_pred_train = classifier.predict(X_train)
cm_train = confusion_matrix(y_pred_train, y_train)

print()
print('Accuracy for training set for SVM = {}'.format((cm_train[0][0] + cm_train[1][1])/len(y_train)))
print('Accuracy for test set for SVM = {}'.format((cm_test[0][0] + cm_test[1][1])/len(y_test)))




```

## 1.4 Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm_test = confusion_matrix(y_pred, y_test)

y_pred_train = classifier.predict(X_train)
cm_train = confusion_matrix(y_pred_train, y_train)

print()
print('Accuracy for training set for Naive Bayes Classifier = {}'.format((cm_train[0][0] + cm_train[1][1])/len(y_train)))
print('Accuracy for test set for Naive Bayes Classifier = {}'.format((cm_test[0][0] + cm_test[1][1])/len(y_test)))
```
